```python
minmax = MinMaxScaler()
onehot = OneHotEncoder()
```

```python
def criandoxy(dados,input_colunas,output_colunas):
    x = dados[input_colunas].copy()
    x_numero = x.select_dtypes(exclude='object')
    x_numero = onehot.fit_transform(x)
    x_minmax = x.selcet_dtypes(include='object')
    x_minmax = minmax.fit_transform(x_minmax)
    x = x_numero.join(x_minmax)
    y = dados[output_colunas].copy()
```
